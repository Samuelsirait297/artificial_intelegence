```python
# Langkah 1: Mengimpor library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Langkah 2: Membuat Data Penjualan Wuling untuk Tahun 2024
data_penjualan = {
    'Bulan': [
        'Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni',
        'Juli', 'Agustus', 'September', 'Oktober'
    ],
    'Unit_Terjual': np.random.randint(100, 500, 10),
    'Harga_Rata_Rata': np.random.randint(200_000_000, 300_000_000, 10)
}

# Langkah 3: Mengubah Dictionary Menjadi DataFrame
df_penjualan = pd.DataFrame(data_penjualan)

print("Data Penjualan Mobil Listrik Wuling 2024:")
print(df_penjualan)

# Langkah 4: Menghitung Total Pendapatan per Bulan
df_penjualan['Pendapatan_Total'] = df_penjualan['Unit_Terjual'] * df_penjualan['Harga_Rata_Rata']

print("\nData Penjualan Mobil Listrik Wuling 2024 dengan Pendapatan Total:")
print(df_penjualan)

# Langkah 5: Statistik Deskriptif
print("\nStatistik Deskriptif Penjualan Mobil Listrik Wuling 2024:")
print(df_penjualan.describe())


# Langkah 6: Visualisasi Data Penjualan
# Grafik 1: Unit Terjual per Bulan
plt.figure(figsize=(10, 5))
plt.plot(df_penjualan['Bulan'], df_penjualan['Unit_Terjual'], marker='o', color='b', label='Unit Terjual')
plt.title('Penjualan Unit Mobil Listrik Wuling per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Unit Terjual')
plt.legend()
plt.grid(True)
plt.show()

# Grafik 2: Pendapatan Total per Bulan
plt.figure(figsize=(10, 5))
plt.bar(df_penjualan['Bulan'], df_penjualan['Pendapatan_Total'], color='green')
plt.title('Pendapatan Total dari Penjualan Mobil Listrik Wuling per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Pendapatan (Rupiah)')
plt.xticks(rotation=45)
plt.show()

# Langkah 7: Analisis Rata-rata Pendapatan
rata_rata_pendapatan = df_penjualan['Pendapatan_Total'].mean()
print(f"\nRata-rata Pendapatan Bulanan: Rp {rata_rata_pendapatan:,.0f}")

# Langkah 8: Analisis Bulan dengan Penjualan Tertinggi dan Terendah
bulan_tertinggi = df_penjualan[df_penjualan['Unit_Terjual'] == df_penjualan['Unit_Terjual'].max()]
bulan_terendah = df_penjualan[df_penjualan['Unit_Terjual'] == df_penjualan['Unit_Terjual'].min()]

print("\nBulan dengan Penjualan Tertinggi:")
print(bulan_tertinggi[['Bulan', 'Unit_Terjual']])

print("\nBulan dengan Penjualan Terendah:")
print(bulan_terendah[['Bulan', 'Unit_Terjual']])

# Langkah 9: Visualisasi Perbandingan Unit Terjual dan Harga Rata-Rata
plt.figure(figsize=(10, 5))
plt.scatter(df_penjualan['Unit_Terjual'], df_penjualan['Harga_Rata_Rata'], color='red')
plt.title('Perbandingan Unit Terjual dan Harga Rata-Rata')
plt.xlabel('Unit Terjual')
plt.ylabel('Harga Rata-Rata (Rupiah)')
plt.grid(True)
plt.show()
```
